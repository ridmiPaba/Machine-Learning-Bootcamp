```python
import numpy as np
import pandas as pd
```

```python
column_names=['user_id','item_id','ratings','timestamp']
```

```python
df= pd.read_csv('u.data',sep='\t',names=column_names)
```

```python
df.head()
```

```python
movie_titles=pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
df= pd.merge(df,movie_titles,on='item_id')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('white')
```

```python
df.groupby('title')['ratings'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['ratings'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['ratings'].mean())
```

```python
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title')['ratings'].count())
```

```python
ratings.head()
```

```python
ratings['num of ratings'].hist(bins=70)
```

```python
ratings['ratings'].hist(bins=70)
```

```python
sns.jointplot(x='ratings',y='num of ratings',data=ratings,alpha=0.5)
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values ='ratings')
```

```python
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_liarliar= moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars= pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head() #this shows how correlated the user ratings of these movies to user ratings of movie 'Star wars'
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation', ascending=False).head()
```

```python
corr_liarliar=pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar=corr_liarliar.join(ratings['num of ratings'])
```

```python
corr_liarliar
```

```python
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
